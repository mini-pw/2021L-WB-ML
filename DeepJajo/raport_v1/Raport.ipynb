# Short description of 2 articles from references

## Article 1 

### Title:
From local explanations to global understanding with explainable AI for trees
https://www.nature.com/articles/s42256-019-0138-9

### What is it about?
It is focused on  tree-based machine learning models and improving their interpretability.

### Why was this article mentioned?
This article presents some methods for explaining tree-based models. The original article is about building a model predicting whether a patient will survive COVID19 or not and the authors were trying to maintain both high performance and interpretability of their models. In order to achieve this goal, they compared different models, as shown in the Supplementary Table 3. The referenced article contains comparisons of different models too and these comparisons prove  that tree-based models can be more interpretable than other models. 
After all, when building a model for a hospital, the model should have rather good performance and interpretability whenever possible and the referenced article proves that simplicity of a model does not necessarily need to drastically decrease  model’s performance.

### What is interesting in this article?
The article shows that despite popularity of tree-based models there are  many improvements to their interpretability that still can be made. 

Publisher: Nature Machine Intelligence
Authors: Lundberg, S. et al.
Published: Nature Machine Intelligence volume  2, pages 56–67(2020)


## Article 2

### Title: XGBoost: A Scalable Tree Boosting System
https://doi.org/10.1145/2939672.2939785

### What is it about?
It introduces a new, open source scalable machine learning algorithm for tree boosting. Authors describe the algorithm in detail, what new cutting-edge features does it implement. They also point that it is widely used among data scientists and can be applied to complex real life problems. 

### Why was this article mentioned?
It was mentioned in the article because the authors used the model that it implements. They wanted to use a high-performance model that is also interpretable. 


### What is interesting in this article?
The article gives both intuition on tree ensemble using simple graphics and formal formulas for in-depth mathematical understanding of gradient tree boosting. Despite introducing new concepts it goes through past developments to give the reader better understanding of what it bases on and what it contributes. The article is elegant and aesthetically pleasing. 

Authors: Tianqi Chen, Carlos Guestrin
Published: KDD '16: Proceedings of the 22nd ACM SIGKDD International Conference on Knowledge Discovery and Data Mining August 2016 Pages 785–794.
