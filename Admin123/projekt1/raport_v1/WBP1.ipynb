# Opis referencji

## 1. Lundberg, S. et al. From local explanations to global understanding with explainable AI for trees. Nat. Mach. Intell. 2, 56–67 (2020)

Artykuł opublikowany w 2020 roku w journalu Nature Machine Intelligence, od niedawna wydawanym przez Nature, skupionym, jak sama nazwa mówi na tematyce sztucznej inteligencji. 

Praca przedstawia algorytm TreeSHAP, pozwalający na szybkie liczenie wartości SHAP dla modeli drzewiastych, oraz sposoby wyjaśniania modeli oparte na wartościach SHAP (m.in. ważność zmiennych liczona jako suma wartości bezwzględnych wartości SHAP). Autorzy pokazują zastosowanie tych metod na problemach z zakresu medycyny. 

Bardzo niejasne dla nas jest, dlaczego autorzy przytoczyli tą pracę. Referencja do niej podana jest po zdaniu: "Finally, for benchmark purposes, the performances of the interpretable model were compared with other standard methods such as random forest and logistic regression". Prawdopodobnie tą pracą autorzy chcą poprzeć to, że lasy losowe i regresja logistyczna to "standardowe metody" - obie z nich były także używane w cytowanej pracy do porównania modeli (m.in. random forest używany był tam w porównaniu, jak dobrze ważność zmiennych liczona na podstawie wartości SHAP radzi sobie z feature selection, w porównaniu z innymi metrykami ważności zmiennych).

## 2. Ridker, P. M. et al. Rosuvastatin to prevent vascular events in men and women with elevated C-reactive protein. N. Engl. J. Med. 359, 2195–2207 (2008)

Artykuł opublikowany w 2008 roku w The New England Journal of Medicine, prestiżowym journalu publikującym artykuły o tematyce medycznej. 

Rosuwastatyna to lek stosowany w hipercholesterolemii. Cytowane badanie bada wpływ jej podawania na ryzyko występowania poważnych problemów w układzie krążenia, takich, jak m.in. zawał serca, u osób nie mających wysokiego cholesterolu, ale za to mających wysoki poziom CRP (C-reactive protein). 
Udowodniona zostaje skuteczność rosuwastatyny w zmniejszeniu ryzyka takich nagłych chorób układu krążenia u badanych. Wskazana zostaje także korelacja wysokiego poziomu CRP z ryzykiem chorób układu krążenia w przyszłości.

W przypadku tego badania także nie do końca rozumiemy dlaczego zostało ono przywołane. Autorzy analizowanej pracy cytują to badanie, aby poprzeć, że wysoki poziom CRP jest markerem mało optymistycznej prognozy u pacjentów z zespołem ostrej niewydolności oddechowej. Jednak w cytowanym badaniu nie ma ani słowa o chorobach układu oddechowego, a w szczególności o zespole ostrej niewydolności oddechowej. 

Co prawdopodobnie autorzy mieli na myśli: Ryzyko śmierci z powodów kardiologicznych oczywiście pogarsza prognozę osób z niewydolnością oddechową (nawet, jak te dwa problemy są od siebie niezależne - tak samo pogarsza prognozę każdego innego). Ponadto jednoczesne choroby oddechowe i kardiologiczne są dodatkowo trudne w leczeniu.
